Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def is_armstrong(x):
    num_str = str(x)
    n = len(num_str)
    total = sum(int(digit) ** n for digit in num_str)
    return total == x
print(is_armstrong(153))
print(is_armstrong(9474))
print(is_armstrong(123)) 
```

Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

```python
a , b , c = map(int , input().split())
delta = (b**2) - 4*a*c
def quadratic_roots_count(a, b, c):
    if delta > 0:
        return 2
    elif delta == 0:
        return 1
    else:
        return 0
if quadratic_roots_count(a, b, c) == 2:
    print("pt 2 nghiem")
elif quadratic_roots_count(a, b, c) == 1:
    print("pt 1 nghiem")
else:
    print("pt vo nghiem")
```

W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)


```python
def count_non_zero_digits(numbers_list):
    total_count = 0 
    for n in numbers_list:
        if n == 0:
            continue
        n = abs(n)
        while n > 0:
            digit = n % 10
            if digit != 0:
                total_count += 1
            n //= 10           
    return total_count
input_data = input("Nhập các số nguyên cách nhau bởi dấu cách: ")
numbers = list(map(int, input_data.split()))
result = count_non_zero_digits(numbers)
print(f"Tổng số lượng các chữ số khác 0 là: {result}")
```

 Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min].


```python
def max_min(a , b , c):
    maxval = max(a, b , c)
    minval = min(a , b , c)
    return (maxval , minval)
a , b , c = map(int , input().split())
ketqua = max_min(a , b , c)
print(ketqua)

```

Viết hàm tính tổng bình phương các số từ 1 đến n

```python
def tong_binh_phuong(n):
    tong = 0
    for i in range(1, n + 1):
        tong += i**2
    return tong
n = int(input("Nhập n: "))
print(f"Tổng bình phương từ 1 đến {n} là: {tong_binh_phuong(n)}")
```
