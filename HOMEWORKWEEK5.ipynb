 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 


```python
n = int(input("Nhập vào số nguyên dương n: "))
chan = 0
le = 0
temp = n 
if temp == 0:
    chan = 1
else:
    while temp > 0:
        chu_so = temp % 10 
        
        if chu_so % 2 == 0: 
            chan += 1
        else:
            le += 1
            
        temp //= 10

print(f"Số {n} có:")
print(f"- {chan} chữ số chẵn")
print(f"- {le} chữ số lẻ")
```

Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số 

```python
a = input("nhap 1 chuoi:")
hoa = 0 
thuong = 0 
so = 0
for kitu in a:
    if kitu.isdigit():
        so += 1
    elif kitu.isupper():
        hoa += 1
    elif kitu.islower():
        thuong += 1
print(f"so ki tu in hoa la:{hoa}")
print(f"so ki tu in thuong la:{thuong}")
print(f"so ki tu so la:{so}")
```

Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành
thường và thường thành hoa

```python
a = input("nhap 1 chuoi:")
ketqua = " "
for kitu in a:
    if kitu.isupper():
        ketqua += kitu.islower()
    elif kitu.islower():
        ketqua += kitu.upper()
print(ketqua)

```

Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 


```python
n = int(input("Nhập vào số nguyên dương n: "))
if n <= 0:
    print(f"{n} không phải là số dạng 2^k")
else:
    temp = n
    while temp % 2 == 0:
        temp //= 2
    if temp == 1:
        print(f"{n} là số dạng 2^k")
    else:
        print(f"{n} không phải là số dạng 2^k")
```
