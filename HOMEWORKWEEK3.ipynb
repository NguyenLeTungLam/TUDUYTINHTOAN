Nhập vào hai số nguyên a, b. Nếu cả hai số a, b đều âm thì in ra "Yes", ngược lại in ra "No". 



```python
a , b = map(int , input().split())
if a < 0 and b < 0:
    print("yes")
elif a > 0 and b > 0:
    print("no")
else:
    print("2 so trai dau")
```

Nhập vào hai chuỗi a, b gồm các chữ cái thường. In ra "True" nếu chuỗi a dài hơn chuỗi b. Ngược lại in ra "False"

```python
a = input("nhap chuoi a:")
b = input("nhap chuoi b:")
if len(a) > len(b):
    print("True")
else:
    print("false")
```

Nếu a, b, c cấu tạo thành được một tam giác, in ra "Yes". Ngược lại in ra "No".

Nhập vào 4 số nguyên. In ra màn hình số nguyên lớn nhất trong 4 số

```python
a , b , c , d = map(int , input("nhap 4 so nguyen:").split())
solonnhat = max(a , b , c , d)
print(solonnhat)
```

Nhập vào một số nguyên dương n là số năm, kiểm tra xem năm đó có phải là năm nhuận hay không. Năm nhuận là năm
chia hết cho 4 nhưng không chia hết cho 100, hoặc chia hết cho 400. Nếu là năm nhuận in ra "Yes", nếu không in ra "No" 

```python
a = int(input("nhap so nam:"))
if (a % 4 == 0 and a % 100 != 0) or (a % 400 == 0):
    print("Yes")
else:
    print("No")
```

Nhập vào hai số thực a và b của phương trình ax + b = 0. Tìm và in ra nghiệm của phương trình, kết quả làm tròn đến số
thập phân thứ 2. In ra "Vô nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng. 


```python
a = int(input("nhap he so a:"))
b = int(input("nhap he so b:"))
if a == 0 and b != 0:
    print("phuong trinh vo nghiem")
elif a == 0 and b ==0:
    print("phuong trinh vo so nghiem")
elif b == 0:
    print("phuong trinh co nghiem duy nhat x = 0")
else:
    x = -b / a
    x = round(x , 2)
    print("phuong trinh co nghiem la:", x)

```

Nhập vào điểm trung bình của một học sinh và in ra học lực của học sinh đó. Xếp loại học lực theo quy tắc:
- >= 8.0: Giỏi
- >= 6.5: Khá
- >= 5.0: Trung bình
- < 5.0: Yếu 


```python
a = float(input("nhap so diem:"))
if a >= 8.0:
    print("hoc luc gioi")
elif a >= 6.5:
    print("hoc luc kha")
elif a >= 5:
    print("hoc luc trung binh")
else:
    print("hoc luc yeu")
```
