Nhập vào hai số nguyên a, b. Nếu cả hai số a, b đều âm thì in ra "Yes", ngược lại in ra "No". 



```python
a , b = map(int , input().split())
if a < 0 and b < 0:
    print("yes")
elif a > 0 and b > 0:
    print("no")
else:
    print("2 so trai dau")
```

chia hết cho 4 nhưng không chia hết cho 100, hoặc chia hết cho 400. Nếu là năm nhuận in ra "Yes", nếu không in ra "No"

Nhập vào điểm trung bình của một học sinh và in ra học lực của học sinh đó. Xếp loại học lực theo quy tắc:
- >= 8.0: Giỏi
- >= 6.5: Khá
- >= 5.0: Trung bình
- < 5.0: Yếu 


```python
a = float(input("nhap so diem:"))
if a >= 8.0:
    print("hoc luc gioi")
elif a >= 6.5:
    print("hoc luc kha")
elif a >= 5:
    print("hoc luc trung binh")
else:
    print("hoc luc yeu")
```

Phân loại tam giác]
Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp
lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay
"Tam giác thường". 


```python

a = int(input("Nhập cạnh a: "))
b = int(input("Nhập cạnh b: "))
c = int(input("Nhập cạnh c: "))
if (a + b > c) and (a + c > b) and (b + c > a):
    if a == b == c:
        print("Tam giác đều")
    elif a == b or b == c or a == c:
        print("Tam giác cân")
    else:
        print("Tam giác thường")
else:
    print("Không phải tam giác")
```

viêt chương trình chia một số m cho n, kết quả thu được làm tròn xuống

```python

m = float(input("Nhập số bị chia m: "))
n = float(input("Nhập số chia n: "))
if n != 0:
    ket_qua = m // n
    print(f"Kết quả {m} chia {n} làm tròn xuống là: {ket_qua}")
else:
    print("Không thể chia cho số 0!")
```
