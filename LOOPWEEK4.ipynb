[SumOfNumBers]
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để
tính tống các số nguyên từ 1 đến n. (n <= 1000)

```python
n = int(input("nhap so nguyen(n <= 1000):"))
tong = 0
for i in range(1 , n + 1):
    tong += i
print(f"Tổng các số từ 1 đến {n} là: {tong}")
```

[PrimeNumber]
VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi.
Kiểm tra số này có phải là số nguyên tố hay không? 


```python
n = 0
while True:
    try:
        n = int(input("nhap so nguyen:"))
        if 0 < n < 100:
            break
        else:
            print("vui long nhap so tu 0 - 100")
    except ValueError:
        print("ban phai nhap 1 so nguyen")
giaithua = 1
for i in range(1 , n + 1):
    giaithua *= i
print(f"ket qua {n}! = {giaithua}")
```
