In các số dương lẻ bé hơn 100 


```python

n = int(input("Nhập một số nguyên n: "))
temp = n
if temp < 0:
    temp = -temp
if temp == 0:
    count = 1
else:
    count = 0
    while temp > 0:
        temp = temp // 10
        count += 1     

print("Số chữ số là:", count)
```

Bài 5: Nhập vào số nguyên dương n. Tính S = 1 + 2 + 3 + 4 + … + n 


```python

n = int(input("Nhập vào số nguyên dương n: "))
S = 0
for i in range(1, n + 1):
    S = S + i  
print(f"Tổng S = 1 + 2 + ... + {n} là: {S}")
```
