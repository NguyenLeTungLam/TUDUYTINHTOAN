trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python

n = int(input("Nhập số hàng n: "))
m = int(input("Nhập số cột m: "))
matrix = []
for i in range(n):
    row = []
    for j in range(m):
        val = int(input(f"Nhập phần tử [{i}][{j}]: "))
        row.append(val)
    matrix.append(row)

print("\nMa trận vừa nhập là:")
for i in range(n):
    for j in range(m):
        print(f"{matrix[i][j]:4d}", end="")
    print()
```

```python

n, m, k = map(int, input().split())
matrix = []
for _ in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)
idx = k - 1
sum_row = sum(matrix[idx]) if 0 <= idx < n else 0

sum_col = sum(matrix[i][idx] for i in range(n)) if 0 <= idx < m else 0

print(f"Tong hang {k}: {sum_row}")
print(f"Tong cot {k}: {sum_col}")
```
