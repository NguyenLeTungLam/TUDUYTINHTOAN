### Bài 1: Tìm số lớn nhất của 1 list

```python
lst = list(map(int , input().split()))
max = lst[0]
for x in lst:
    if x > max:
        max = x
print(max)
```

### Bài 2: Tìm số nhỏ nhất của 1 list

```python
lst = list(map(int , input().split()))
min = lst[0]
for x in lst:
    if x < min:
        min = x
print(min)
```

### Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list. Các số khác không thay đổi thứ tự trước sau
3 4 6 7 1 9 0 0 23 0 0 34 67 89
0 0 0 0 3 4 6 7 1 9 23 34 67 89

```python
lst = [3, 4, 6, 7, 1 ,9, 0 ,0 ,23 ,0, 0 ,34, 67, 89]
zero = []
others = []
for i in lst:
    if i == 0:
        zero.append(i)
    else:
        others.append(i)
result = zero + others
print(result)
```

### Bài 6: Xóa tất cả các số giống nhau trong list, Các số khác giữ nguyên vị trí trước sau

```python
lst = list(map(int , input().split()))
result = []
for x in lst:
    if x not in result:
     result.append(x)
print(result)
```

### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang
9 8 7 6 5 4 0: Giam
1 2 3 7 1 0 2: Vo danh
1 2 2 4 5 7: Vo danh

```python
lst = list(map(int , input().split()))
if lst == sorted(lst):
    print("day tang")
elif lst == sorted(lst , reverse = True):
    print("day giam")
else:
    print("vo dinh")
```
